```python
import pandas
import math
%matplotlib inline

country = pandas.read_csv("col/distribution.txt", sep='\t')
country
```

```python
reference = pandas.read_csv("col/newreference.txt", sep='\t')
country = country.set_index("taxonID")
#reference = reference.set_index("taxonID")


```

```python

all_countries = set()
for i in country["locality"]:
    #print(i)
    all_countries.add(i)
```

```python
all_countries

```

```python
arr = [0] * 20
usedID = set()
discoveries = pandas.DataFrame(index = list(range(2020)), columns=list(all_countries), dtype=int).fillna(0)
for i in reference.index:
    
    if (reference.taxonID[i] in usedID):
        continue
        
    if (len(usedID) % 5000 == 0):
        print(len(usedID))
    usedID.add(reference.taxonID[i])
    try:
        current_countries = country.loc[reference.taxonID[i], "locality"]
        #print(country.loc[reference.taxonID[i], "description"])
        year = int(reference.date[i])
    except:
        #print(reference.taxonID[i])
        continue
        
    if year > 2018:
        continue
    #if (year not in discoveries.index):
   #     discoveries.loc[year] =
        #if (c not in discoveries.columns):
        #    discoveries[c] = pandas.Series([0] * 2020, index = discoveries.index)
        #print(discoveries)
    #print(current_countries)
    if type(current_countries) is str:
        discoveries[current_countries][year] += 1
    else:
        for c in current_countries:
            discoveries[c][year] += 1
    #print(i)
#new_species = reference[(reference["date"] > 2000) & (reference["type"] == "taxon")]
discoveries
#print(new_species.loc[74390, :])
```

```python
i = 1
while (sum(discoveries.loc[i, :]) == 0):
    discoveries = discoveries.drop(i, axis=0)
    i += 1
```
